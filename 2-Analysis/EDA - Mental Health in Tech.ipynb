### Methodology

For my exploratory analysis, I will perform some descriptive analysis using the 2019 survey results to answer my research questions. I will make different plots and graphs to summarize the characteristics of the dataset pertinent to the questions. 

A bar chart to study the perception of participants across the the different score levels (1 to 5)
- Overall, how well do you think the tech industry supports employees with mental health issues?

A funnel visualization to see what percent of total participants answered 'Yes' at the following 4 stages:
- Coverage: Does your employer provide mental health benefits as part of healthcare coverage? 
- Awareness: Do you know the options for mental health care available under your employer-provided health coverage?"
- Discussions: Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?
- Help Resources: Does your employer offer resources to learn more about mental health disorders and options for seeking help?

Pie charts/bar garphs to study the distribution of participants for different levels of difficulty in leave policies
- If a mental health issue prompted you to request a medical leave from work, how easy or difficult would it be to ask for that leave?

Grouped bar plots to compare the number of participants for each score level (0 to 10) for the following questions:
- Overall, how much importance does your employer place on physical health?	
- Overall, how much importance does your employer place on mental health?

Bar plots to compare the responses of participants to the question "Do you *currently* have a mental health disorder?"  and "Have you had a mental health disorder in the past?" across age, gender, race, country of work, and country of residence.
- What is your age?	
- What is your gender?	
- What is your race?	
- What country do you *live* in?	
- What country do you *work* in?	

Scatter plot to compare if men are less expressive about mental health than women
- Have you ever discussed your mental health with coworkers?
- What is your gender?
